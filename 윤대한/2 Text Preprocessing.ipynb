#2-1 Tokenization (토큰화)  
Tokenization이란 주어진 corpus에서 token이라 불리는 단위로 나누는 작업을 뜻 함.  

token의 단위는 상황에 따라 다름  
이 장에서 여러가지 상황으로 Tokenization에 대한 개념을 이해할 것임.

##1. 단어 토큰화(Word Tokenization)
token의 단위를 word로 하는 경우.  
이때 word는 단어 단위 외에도 **단어구, 의미를 갖는 문자열**로도 간주함.


예로는 구두점(punctuation)과 같은 문자는 제외하고, 단어들만 tokenization 함.

##2. 토큰화 중 생기는 선택의 순간
tokenization의 기준을 생각해봐야 하는 경우가 발생함.  
예로 들면 ``아포스트로피(`)``가 있음  

토큰화 과정을 예시를 통해 알아볼 예정  

**NLTK**는 영어 코퍼스를 토큰화하기 위한 도구를 제공함.  
`nltk.tokenize`에는 `word_tokenize`와 `WordPunctTokenizer`가 있음.  
이 두개로 어떻게 토큰화를 하는지 알아보자.

````python
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
````

````python
print('단어 토큰화1 :',word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
print('단어 토큰화2 :',WordPunctTokenizer().tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
````

일단 둘의 차이를 보자.  
`word_tokenize`는 Don\`t를 Do와 n\`t로 분리 하였으나  
Jone\`s는 Jone과 \`s로 분리함.  
`WordPunctTokenizer`은 Don\`t를 Don과 n\`t로 분리 하였으나  
Jone\`s는 Jone과 \`와 s로 분리함.  

이 둘의 차이점은 확실하다.  

keras에서도 토큰화 도구인 `text_to_word_sequence`를 지원함.



````python
from tensorflow.keras.preprocessing.text import text_to_word_sequence
print('단어 토큰화3 :',text_to_word_sequence("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
````

`text_to_word_sequence`는 Don\`t와 Jone\`s를 그대로 토큰화 함.  
하지만 전부 소문자화 하였음.  
구두점도 전부 삭제 되었다.  



##3. 토큰화에서 고려해야할 사항
단순희 구두점과 공백을 기준으로 잘라내는 작업이라 보기 어려움.  

###1) 구두점이나 특수 문자를 단순 제외해서는 안된다.
날짜나 소수점이 있는 숫자는 기호를 없애면 의미가 달라짐  

###2) 줄임말과 단어 내에 띄어쓰기가 있는 경우.
띄어쓰기가 있어도 하나의 토큰으로 봐야하는 경우도 있을 수 있음.

###3) 표준 토큰화 예제
**Penn Treebank Tokenization**  
rule 1. 하이푼으로 구성된 단어는 하나로 유지.  
rule 2. doesn`t와 같이 아포스트로피로 '접어'가 함께하는 단어는 분리해준다.  

````python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

text = "Starting a home-based restaurant may be an ideal. it doesn't have a food chain or restaurant of their own."
print('트리뱅크 워드토크나이저 :',tokenizer.tokenize(text))
#dose + n`t   home-based
````

#2-4 Stopword (불용어)
필요없는 단어 제거

````python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt
stop_words_list = stopwords.words('english')
print('불용어 개수 :', len(stop_words_list))
print('불용어 10개 출력 :',stop_words_list[:10])
````

````python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []
for word in word_tokens:
    if word not in stop_words:
        result.append(word)

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
````

한국어에서 불용어 제거

````python
okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
````

#2-6 Integer Encoding (정수 인코딩)

##1. 정수 인코딩(Integer Encoding)
빈도수가 높은 순서대로 정수를 부여하는 방법

###dictionary 사용

````python
import nltk
nltk.download()
````

````python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
````

````python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
````

````python
#문장 토큰화
sentences = sent_tokenize(raw_text)
print(sentences)
````

````python
#문장 정제 및 정규화작업
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
  tokenized_sentence = word_tokenize(sentence)
  result = []

  for word in tokenized_sentence:
    word = word.lower() #소문자로 줄이기
    if word not in stop_words:
      if len(word) > 2:
        result.append(word)
        if word not in vocab:
          vocab[word] = 0
        vocab[word]+=1
  preprocessed_sentences.append(result)
print(preprocessed_sentences)
````

````python
print("단어 집합 :",vocab)
````

````python
print(vocab["barber"])
````

````python
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sorted)
````

````python
# 빈도가 높은 단어부터 낮은 정수 부여
word_to_index = {}
i = 0
for(word,frequency)in vocab_sorted:
  if frequency>1:
    i+=1
    word_to_index[word]=i
print(word_to_index)
````

````python
# 상위 단어들만 추출

vocab_size = 5
words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1]

for w in words_frequency:
  del word_to_index[w]
print(word_to_index)
````

````python
# OOV 문제 (단어 집합에 없는 단어)
word_to_index['OOV'] = len(word_to_index) + 1
print(word_to_index)
````

````python
# 정수 인코딩
encoded_sentences = []
for sentence in preprocessed_sentences:
  encoded_sentence = []
  for word in sentence:
    try:
      encoded_sentence.append(word_to_index[word])
    except KeyError:
      encoded_sentence.append(word_to_index['OOV'])
  encoded_sentences.append(encoded_sentence)
print(encoded_sentences)
````
