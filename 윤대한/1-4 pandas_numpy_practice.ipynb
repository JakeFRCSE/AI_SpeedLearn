```python
import pandas as pd
```

```python
sr = pd.Series([17000, 18000, 1000, 5000], index=["피자", "치킨", "콜라", "맥주"])
print('시리즈 출력 :')
print('-'*15)
print(sr)
```

```python
print('시리즈의 값 : {}'.format(sr.values))
print('시리즈의 인덱스 : {}'.format(sr.index))
```

```python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df = pd.DataFrame(values, index=index, columns=columns)

print('데이터프레임 출력 :')
print('-'*18)
print(df)
```

```python
print('데이터프레임의 인덱스 : {}'.format(df.index))
print('데이터프레임의 열이름: {}'.format(df.columns))
print('데이터프레임의 값 :')
print('-'*18)
print(df.values)
```

```python
mat = np.array([[1,2,3],[4,5,6]])
print(mat)
```

```python
slicing_mat = mat[0, :] # :(콜론)은 전체라는 뜻 앞뒤에 숫자가 있으면 거기부터 or 거기 까지 라는 뜻
print(slicing_mat)
```

```python
slicing_mat = mat[:,1]
print(slicing_mat)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.show()
```

```python
#없을 경우 점프함.
plt.title('test')
plt.plot([1,2,4,5],[2,4,8,6])
plt.show()
#순서대로 선긋기
plt.title('test')
plt.plot([1,2,5,4,2],[2,4,8,6,1])
plt.show()
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.title('students')
plt.plot([1,2,3,4],[2,4,8,6])
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10])
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student'])
plt.show()
```
